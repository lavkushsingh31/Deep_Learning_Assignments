# Artificial Neural Network Modelling on Synthetic Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_moons, make_circles
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### Generating synthetic datasets

##### Moons Dataset

```python
X_moons,y_moons = make_moons(n_samples=5000, noise=0.1, random_state=101)
```

```python
X_moons.shape, y_moons.shape
```

```python
plt.figure(figsize=(7,4))
plt.scatter(X_moons[:, 0], X_moons[:,1], c = y_moons)
plt.show()
```

##### Circles Dataset

```python
X_circles,y_circles = make_circles(n_samples=5000, noise=0.1, factor=0.4, random_state=101)
```

```python
X_circles.shape, y_circles.shape
```

```python
plt.figure(figsize=(7,4))
plt.scatter(X_circles[:, 0], X_circles[:,1], c = y_circles)
plt.show()
```

##### Classification Dataset

```python
X_clf,y_clf = make_classification(n_samples=5000, n_features=2, n_redundant=0, random_state=101)
```

```python
X_clf.shape, y_clf.shape
```

```python
plt.figure(figsize=(7,4))
plt.scatter(X_clf[:, 0], X_clf[:,1], c = y_clf)
plt.show()
```

### Model Building

##### Initializing and training the model - Moons Dataset

```python
model = Sequential(name='ANN_for_Moons_Dataset_Classification')

model.add(Dense(name = "Dense_Layer_1", units=16, activation='relu', input_shape = (2,)))
model.add(Dense(name = "Dense_Layer_2", units=8,activation='relu'))
model.add(Dense(name = "Dense_Layer_3", units=4,activation='relu'))
model.add(Dense(name = "Output_Layer", units=1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics='accuracy')
model.summary()
```

```python
model.fit(x=X_moons,
          y=y_moons,
          batch_size=1000,
          epochs=500,
          validation_split=0.1)
```

```python
model = Sequential(name='ANN_for_Circles_Dataset_Classification')

model.add(Dense(name = "Dense_Layer_1", units=16, activation='relu', input_shape = (2,)))
model.add(Dense(name = "Dense_Layer_2", units=8,activation='relu'))
model.add(Dense(name = "Dense_Layer_3", units=4,activation='relu'))
model.add(Dense(name = "Output_Layer", units=1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics='accuracy')
model.summary()
```

```python
model.fit(x=X_circles,
          y=y_circles,
          batch_size=1000,
          epochs=500,
          validation_split=0.1)
```

```python
model = Sequential(name='ANN_for_Classification_Dataset_Classification')

model.add(Dense(name = "Dense_Layer_1", units=16, activation='relu', input_shape = (2,)))
model.add(Dense(name = "Dense_Layer_2", units=8,activation='relu'))
model.add(Dense(name = "Dense_Layer_3", units=4,activation='relu'))
model.add(Dense(name = "Output_Layer", units=1, activation='sigmoid'))
```
